### Exercise 1
Given the maximization function of $T+1$ periods, we can write the maximization problem of $T=1$ as follows: $$\max_{W_{T+1}} \ u(W_T - W_{T+1})$$ Given $W_T$, if we want to maximize this utility (and the utility function is strictly increasing), we need to set $W_{T+1} = \min (W_{T+1})$. Since we have constraints on the cake size that $W_{T} \geq 0$ for $\forall \ T$, $W_{T+1}=0$. 

### Exercise 2

If the individual lives for two periods $T=2$, the maximization problem for next period $W_3$ in period 2 is $$\max_{W_3 \in \ [0, W_2]} \ u(W_2-W_3)$$ 
Given the solutions in Exercise 1, if we want to maximizes utility in period 2, we need to set $W_3 = \min(W_3)$, which is 0. <br> 
Similarly, the maximization problem in period 2 can be expressed as the following: 
$$\max_{W_2 \in \ [0, W_1],\ W_3 \in \ [0, W_2]} \ u(W_1-W_2) + \beta \ u(W_2-W_3)$$
Since $W_3 = 0$, we can simplify the equation above as:
$$\max_{W_2 \in \ [0, W_1]} \ u(W_1-W_2) + \beta \ u(W_2)$$

### Exercise 4

Given the envelop theorem and the policy function for $W_T$ period is $W_T = \psi (W_{T-1})$, the value function $V_{T-1}$ can be written as:
$$V_{T-1}(W_{T-1}) \equiv \max_{W_T} u(W_{T-1} - W_T)+ \beta\ V_{T}(W_T)$$
$$\Rightarrow V_{T-1}(W_{T-1}) \equiv \max_{\psi (W_{T-1})} u\big(W_{T-1} - \psi (W_{T-1})\big)+ \beta\ V_{T}\big(\psi (W_{T-1})\big)$$
The first order condition of this question is
$$-u'(W_{T-1} - W_T) + \beta\ u'(W_T) = 0$$
$$\Rightarrow -u'\big(W_{T-1} - \psi (W_{T-1})\big) + \beta\ u'\big(\psi (W_{T-1})\big) = 0$$

### Exercise 5

Since the utility function is written as $u(c) = \ln(c)$, given the arguments above, we know
$$V_{T}(\overline{W}) = \ln(\overline{W})$$
and since $T$ is the last period, we have $$\psi_T(W_T) = W_{T+1} = 0$$
Given the characterization of $V_{T-1}$ in the exercise above, we have
$$V_{T-1}(\overline{W}) = \max_{W_{T+1}} \ln(\overline{W} - W_{T+1}) + \beta \ln(W_{T+1})$$
Solving for first order condition of the maximization problem above, we have
$$-\frac{1}{\overline{W} - W_{T+1}} + \beta \frac{1}{W_{T+1}} = 0$$
$$\Rightarrow \psi_{T-1}(W_{T-1}) = W_{T+1} = \frac{\beta}{\beta +1} \overline{W} \neq \psi_T(W_T)$$
If we plug in the equation above to $V_{T-1}(\overline{W})$, we got
$$V_{T-1}(\overline{W}) = \ln(\frac{\overline{W}}{\beta +1}) + \beta \ln (\frac{\beta \ \overline{W}}{\beta +1} ) \neq V_{T}(\overline{W})$$
Therefore, we show that the value function and policy function in period $T$ and $T-1$ are not equal.

### Exercise 6

The Bellman Equation in period $T-2$ can be written as the following:
$$V_{T-2}(W_{T-2}) = \max_{W_{T-1}, W_{T}} \ln(W_{T-2}-W_{T-1}) + \beta \ln(W_{T-1} - W_T) +\beta^2 \ln(W_T)$$
Given the solutions obtained in exercise 5 that $W_{T} = \psi_{T-1}(W_{T-1}) = \frac{\beta}{\beta+1} W_{T-1}$, we can simplify the equation above as the following:
$$V_{T-2}(W_{T-2}) = \max_{W_{T-1}} \ln(W_{T-2}-W_{T-1}) + \beta \ln(\frac{W_{T-1}}{1+\beta}) +\beta^2 \ln(\frac{\beta}{1+\beta}W_{T-1})$$
The first order condition characterize the policy function in period $T-2$ can be written as:
$$\frac{1}{W_{T-2}-W_{T-1}} = \frac{\beta^2+\beta}{W_{T-1}}$$
$$\Rightarrow \psi_{T-2}(W_{T-2}) = W_{T-1} = \frac{\beta^2+\beta}{\beta^2+\beta+1}W_{T-2}$$
Therefore,
$$V_{T-2}(W_{T-2}) = \ln(\frac{W_{T-2}}{\beta^2+\beta+1}) + \beta \ln(\frac{\beta}{\beta^2+\beta+1}W_{T-2}) + \beta^2 \ln(\frac{\beta^2}{\beta^2+\beta+1}W_{T-2})$$

### Exercise 7

By induction, we can infer that the policy function in $T-s$ period is
$$\psi_{T-s}(W_{T-s}) = W_{T-s+1} = \frac{\beta^s + \beta^{s-1}+ \ldots +\beta}{\beta^s + \beta^{s-1}+ \ldots +\beta+1}W_{T-s}$$
and the value function is
$$V_{T-s}(W_{T-s}) = \ln(\frac{1}{M}W_{T-s}) + \beta \ln(\frac{\beta}{M}W_{T-s}) + \ldots + \beta^s \ln(\frac{\beta^s}{M}W_{T-s})$$
where $M = \beta^s + \beta^{s-1}+ \ldots +\beta+1$ <br>
As $s \rightarrow \infty$,
$$\lim_{s \to \infty} \psi_{T-s}(W_{T-s}) = \beta\ W_{T-s} = \psi(W_{T-s})$$
$$\lim_{s \to \infty} V_{T-s}(W_{T-s}) =  \left(\frac{1}{1-\beta}\right) \ln \big((1-\beta){ W_{T-s}}\big)+\frac{\beta}{(1-\beta)^{2}} \ln \beta = V(W_{T-s})$$

```python
from scipy.stats import norm
from scipy.integrate import quad
import pandas as pd

#The function of using Newton-Cotes method to genderate probability
def NCnorm(mu, sigma, N, k):
    z_points = np.linspace(mu-k*sigma, mu+k*sigma, N)
    w_points = np.zeros(N)
    for i in range(N):
        if i==0:
            w_points[i] = norm.cdf((z_points[0]+z_points[1])/2, loc=mu, scale=sigma)
        elif i==N-1:
            w_points[i] = 1 - norm.cdf((z_points[N-2]+z_points[N-1])/2, loc=mu, scale=sigma)
        else:
            pdf = lambda x: norm.pdf(x, loc=mu, scale=sigma)
            z_min = (z_points[i-1]+z_points[i])/2
            z_max = (z_points[i]+z_points[i+1])/2
            w_points[i] = quad(pdf, z_min, z_max)[0]
    return z_points, w_points

sigma = 0.5
mu = 4*sigma
M=7
z_nodes, w_nodes = NCnorm(mu, sigma, M, k=3)
ncdf = pd.DataFrame({"z nodes": z_nodes, "probability": w_nodes})
ncdf
```
